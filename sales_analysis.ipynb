```python
duplicate=df.duplicated().sum()   #Finding duplicate values int the dataset and summing them.
print("Number of duplicate rows:", duplicate)
df.drop_duplicates(inplace=True)   #Dropping duplicate values from the dataset (if any)
```

```python
df["Order Date"]
df["Ship Date"]
```

**Fixing date column and Creating Year and Month features for time-based sales analysis.**

```python
#For Order Date Column.
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month
df["Order Month Name"] = df["Order Date"].dt.month_name()
df["Order Day"] = df["Order Date"].dt.day
df["Order Day Name"] = df["Order Date"].dt.day_name()

#For Ship Date Column.
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

df["Ship Year"] = df["Ship Date"].dt.year
df["Ship Month"] = df["Ship Date"].dt.month
df["Ship Month Name"] = df["Ship Date"].dt.month_name()
df["Ship Day"] = df["Ship Date"].dt.day
df["Ship Day Name"] = df["Ship Date"].dt.day_name()
```

**Showing the Dataset after creating Year, Month and Day Features**

```python
df.head()
```

**üí∞ Total Sales and Total Profit**

**üåç Sales by Region**

Analyzing sales distribution across different regions to identify the highest revenue-generating areas.

```python
df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
```

Here in this Output it shows that **West Region** generates the highest revenue

**üßæ Sales by Category**

Evaluating which product categories contribute the most to overall revenue.



```python
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

In the Output we can see that **Office Supplies** have the highest sales.

**Now here are the *Top 10 Products* by Sales** 

```python
df.groupby("Product Name")["Sales"].sum().nlargest(10)
```

**Now we will Visualize the Data**

**The given bar graph gives Profit by Category**

```python
df.groupby("Category")["Profit"].sum().plot(kind="bar", title="Profit By Category", color="teal")
```

**The given chart give the Sales Distribution by Category**

```python
# Sales Distribution by Category

Category_Sales = df.groupby("Category")["Sales"].sum().nlargest(10)

plt.figure(figsize=(5,5))
plt.pie(labels=Category_Sales.index, x=Category_Sales.values, autopct="%1.3f%%", startangle=90)
plt.title("Sales Distribution by Category")
plt.show()
```

**The given chart highlights the top 10 revenue-generating products.**

```python
Top_10_Products = df.groupby("Product Name")["Sales"].sum().nlargest(10)
plt.figure(figsize=(10,6))
Top_10_Products.plot(kind="barh", color="coral")
plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product Name")
```

**üìä Sales vs Profit**

Analyzing the relationship between sales and profit to identify high-revenue but low-profit transactions.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Sales", y="Profit", color="Green", data=df)
plt.title("Sales vs Profit")
```

**Loss Making Porducts**

```python
Loss_df = df[df["Profit"] < 0]
Loss_df.head()
```

**Now we will visualize Top 10 loss making products**

```python
Top_loss_products = Loss_df.groupby("Product Name")["Profit"].sum().nsmallest(10).sort_values()

plt.figure(figsize=(10,6))
Top_loss_products.plot(kind="barh", color="red")
plt.title("Top 10 Loss-Making Products")
plt.xlabel("Profit")
plt.ylabel("Product Name")
plt.show()
```
